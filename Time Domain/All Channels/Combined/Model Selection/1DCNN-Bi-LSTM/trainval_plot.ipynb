```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of CSV files
csv_files = [f'epoch_{i}.csv' for i in range(1, 11)]

# Initialize lists to store data
epochs = []
training_accuracy = []
training_loss = []
validation_accuracy = []
validation_loss = []

# Read data from each CSV file
for file in csv_files:
    df = pd.read_csv(file)
    epochs.append(df['Epoch'][:100])
    training_accuracy.append(df['Training Accuracy'][:100])
    training_loss.append(df['Training Loss'][:100])
    validation_accuracy.append(df['Validation Accuracy'][:100])
    validation_loss.append(df['Validation Loss'][:100])

# Convert lists to DataFrames
epochs_df = pd.concat(epochs, axis=1)
training_accuracy_df = pd.concat(training_accuracy, axis=1)
training_loss_df = pd.concat(training_loss, axis=1)
validation_accuracy_df = pd.concat(validation_accuracy, axis=1)
validation_loss_df = pd.concat(validation_loss, axis=1)

# Calculate mean and standard deviation
mean_training_accuracy = training_accuracy_df.mean(axis=1)
std_training_accuracy = training_accuracy_df.std(axis=1)
mean_training_loss = training_loss_df.mean(axis=1)
std_training_loss = training_loss_df.std(axis=1)
mean_validation_accuracy = validation_accuracy_df.mean(axis=1)
std_validation_accuracy = validation_accuracy_df.std(axis=1)
mean_validation_loss = validation_loss_df.mean(axis=1)
std_validation_loss = validation_loss_df.std(axis=1)

# Plotting
epochs_range = range(1, 101)

plt.figure(figsize=(8, 6))

# Training Accuracy
plt.subplot(2, 2, 1)
plt.plot(epochs_range, mean_training_accuracy, label='Mean Training Accuracy', color='blue', lw=1)  # Thicker line
plt.fill_between(epochs_range, mean_training_accuracy - std_training_accuracy, mean_training_accuracy + std_training_accuracy, color='blue', alpha=0.2)
# plt.title('Training Accuracy', fontweight='bold', fontsize=14)
plt.xlabel('Epochs', fontweight='bold', fontsize=12)
# plt.ylabel('Training Accuracy', fontweight='bold', fontsize=12)
plt.legend(fontsize=12, loc='best', prop={'weight': 'bold'})

# Training Loss
plt.subplot(2, 2, 2)
plt.plot(epochs_range, mean_training_loss, label='Mean Training Loss', color='red', lw=1)  # Thicker line
plt.fill_between(epochs_range, mean_training_loss - std_training_loss, mean_training_loss + std_training_loss, color='red', alpha=0.2)
# plt.title('Training Loss', fontweight='bold', fontsize=14)
plt.xlabel('Epochs', fontweight='bold', fontsize=12)
# plt.ylabel('Training Loss', fontweight='bold', fontsize=12)
plt.legend(fontsize=12, loc='best', prop={'weight': 'bold'})

# Validation Accuracy
plt.subplot(2, 2, 3)
plt.plot(epochs_range, mean_validation_accuracy, label='Mean Validation Accuracy', color='green', lw=1)  # Thicker line
plt.fill_between(epochs_range, mean_validation_accuracy - std_validation_accuracy, mean_validation_accuracy + std_validation_accuracy, color='green', alpha=0.2)
# plt.title('Validation Accuracy', fontweight='bold', fontsize=14)
plt.xlabel('Epochs', fontweight='bold', fontsize=12)
# plt.ylabel('Validation Accuracy', fontweight='bold', fontsize=12)
plt.legend(fontsize=12, loc='best', prop={'weight': 'bold'})

# Validation Loss
plt.subplot(2, 2, 4)
plt.plot(epochs_range, mean_validation_loss, label='Mean Validation Loss', color='purple', lw=1)  # Thicker line
plt.fill_between(epochs_range, mean_validation_loss - std_validation_loss, mean_validation_loss + std_validation_loss, color='purple', alpha=0.2)
# plt.title('Validation Loss', fontweight='bold', fontsize=14)
plt.xlabel('Epochs', fontweight='bold', fontsize=12)
# plt.ylabel('Validation Loss', fontweight='bold', fontsize=12)
plt.legend(fontsize=12, loc='best', prop={'weight': 'bold'})

# Adjust layout
plt.tight_layout()

# # Save the figure as an image file
# plt.savefig('training_performance.png', bbox_inches='tight', dpi=1200)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# List of CSV files
csv_files = [f'epoch_{i}.csv' for i in range(1, 11)]

# Initialize lists to store data
epochs = []
training_accuracy = []
training_loss = []
validation_accuracy = []
validation_loss = []

# Read data from each CSV file
for file in csv_files:
    df = pd.read_csv(file)
    epochs.append(df['Epoch'][:100])
    training_accuracy.append(df['Training Accuracy'][:100])
    training_loss.append(df['Training Loss'][:100])
    validation_accuracy.append(df['Validation Accuracy'][:100])
    validation_loss.append(df['Validation Loss'][:100])

# Convert lists to DataFrames
epochs_df = pd.concat(epochs, axis=1)
training_accuracy_df = pd.concat(training_accuracy, axis=1)
training_loss_df = pd.concat(training_loss, axis=1)
validation_accuracy_df = pd.concat(validation_accuracy, axis=1)
validation_loss_df = pd.concat(validation_loss, axis=1)

# Calculate mean and standard deviation
mean_training_accuracy = training_accuracy_df.mean(axis=1)
std_training_accuracy = training_accuracy_df.std(axis=1)
mean_training_loss = training_loss_df.mean(axis=1)
std_training_loss = training_loss_df.std(axis=1)
mean_validation_accuracy = validation_accuracy_df.mean(axis=1)
std_validation_accuracy = validation_accuracy_df.std(axis=1)
mean_validation_loss = validation_loss_df.mean(axis=1)
std_validation_loss = validation_loss_df.std(axis=1)

# Plotting
epochs_range = range(1, 101)

plt.figure(figsize=(6, 4))

# Training and Validation Accuracy (Overlapped)
plt.subplot(2, 1, 1)
plt.plot(epochs_range, mean_training_accuracy, label='Training Accuracy', color='blue', lw=2)  # Thicker line
plt.fill_between(epochs_range, mean_training_accuracy - std_training_accuracy, mean_training_accuracy + std_training_accuracy, color='blue', alpha=0.2)
plt.plot(epochs_range, mean_validation_accuracy, label='Validation Accuracy', color='green', lw=2)  # Thicker line
plt.fill_between(epochs_range, mean_validation_accuracy - std_validation_accuracy, mean_validation_accuracy + std_validation_accuracy, color='green', alpha=0.2)
plt.xlabel('Epochs', fontweight='bold', fontsize=12)
plt.ylabel('Accuracy', fontweight='bold', fontsize=12)
plt.legend(fontsize=12, loc='best', prop={'weight': 'bold'})

# Training and Validation Loss (Overlapped)
plt.subplot(2, 1, 2)
plt.plot(epochs_range, mean_training_loss, label='Training Loss', color='red', lw=2)  # Thicker line
plt.fill_between(epochs_range, mean_training_loss - std_training_loss, mean_training_loss + std_training_loss, color='red', alpha=0.2)
plt.plot(epochs_range, mean_validation_loss, label='Validation Loss', color='purple', lw=2)  # Thicker line
plt.fill_between(epochs_range, mean_validation_loss - std_validation_loss, mean_validation_loss + std_validation_loss, color='purple', alpha=0.2)
plt.xlabel('Epochs', fontweight='bold', fontsize=12)
plt.ylabel('Loss', fontweight='bold', fontsize=12)
plt.legend(fontsize=12, loc='best', prop={'weight': 'bold'})

# Adjust layout
plt.tight_layout()

# Display the plot
plt.show()
```
